**Write a python code to check whether a number is prime or not.**

```python
def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False

    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6

    return True


number = int(input("Enter a number: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
def is_palindrome(word):
    """Check if a word is a palindrome."""
    return word == word[::-1]

def find_palindromes_in_text(text):
    """Process a string and return palindrome words with their counts."""
    palindrome_counts = {}

    words = text.split()
    for word in words:

        clean_word = ''.join(char for char in word if char.isalnum()).lower()

        if clean_word and is_palindrome(clean_word):
            if clean_word in palindrome_counts:
                palindrome_counts[clean_word] += 1
            else:
                palindrome_counts[clean_word] = 1

    return palindrome_counts

text = input("Enter the text: ")
palindromes = find_palindromes_in_text(text)

if palindromes:
    print("Palindrome words and their counts:")
    for word, count in palindromes.items():
        print(f"{word}: {count}")
else:
    print("No palindromes found in the text.")
```
